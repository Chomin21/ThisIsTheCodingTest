```python
#백준 입력용
#풀었지만 시간초과
#not in queue쪽에서 시간초과인 것 같지만 내 논리상 이 코드를 고치긴 어려움

from collections import deque

N, M, K, X = map(int, input().split())

road = []
dist = [0]*N

for i in range(M):
    road.append(list(map(int, input().split())))
    
def bfs(X):
    city = []
    queue = deque()
    queue.append(X)
    distance = 0
    
    while queue:
        x = queue.popleft()
        distance += 1
        for k in road:
            if x == k[0] and k[1] not in queue:
                queue.append(k[1])
                dist[k[1]-1] = dist[k[0]-1] + 1
                if dist[k[1]-1] == K:
                    city.append(k[1])
    if len(city) == 0:
        return -1
    else:
        city.sort()
        return city
    
answer = bfs(X)
for i in answer:
    print(i)
```

```python
#윤서안 코드

# 특도찾
import sys
from collections import deque

input = sys.stdin.readline

n, m, k, x = map(int, input().split())
graph = [[] for _ in range(n + 1)] #인접리스트 선언

for _ in range(m):
    a, b = list(map(int, input().split()))
    graph[a].append(b)

city = [-1] * (n + 1) #모든 도시들의 거리를 -1로 초기화. visited의 역할을 함.
city[x] = 0 #시작 도시는 0으로 초기화

queue = deque([x]) #시작도시를 queue에 넣음
while queue:
    v = queue.popleft() #queue에 있는 요소들을 앞에서부터 뽑음
    
    for i in graph[v]: #v와 연결된 도시들을 하나씩 가져와서 확인
        if city[i] == -1: #방문하지 않은 도시인 경우. (지금 방문하고 있는 도시가 i가 되는 것)
            city[i] = city[v] + 1 #현재도시(v)의 거리 + 1을 도시 i에 넣어줌.
            queue.append(i)
        
#특정 거리를 찾은 경우 출력
for i in range(1, n + 1):
    if city[i] == k:
        print(i)
        
#거리가 도시중에 없는 경우
if k not in city:
    print(-1)
```
