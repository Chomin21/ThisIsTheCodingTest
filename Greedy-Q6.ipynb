```python
#무지의 먹방 라이브
#내 최종 답안
#정확성 100(42.9), 효율성 0

def solution(food_times, k):
    time = 0
    ind = 0
    
    while time < k+1:
        #answer(인덱스)가 유효한 값 범위 내인지 확인
        while True:
            if ind == len(food_times):
                ind = 0
            #다음값이 0이 아닌 경우 반복문 종료
            if food_times[ind] != 0:
                break
            #food_times안 값이 모두 0이면 -1 출력
            if any(food_times) == False:
                answer = -1
                return answer
            ind += 1
        
        food_times[ind] -= 1
        answer = ind + 1
    
        ind += 1
        time += 1
    return answer
        


food_times = list(map(int, input().split()))
k = int(input())
print(solution(food_times, k))
```
