```python
#부품 찾기
#my solution
#sys는 주피터에서 잘 작동하지 않아서 pycharm에서 코딩 후 복붙
import sys

N = int(sys.stdin.readline())
input_n = list(map(int, sys.stdin.readline().split()))
input_n.sort()

M = int(sys.stdin.readline())
input_m = list(map(int, sys.stdin.readline().split()))

#input_m에 있는 것들을 순서대로 하나씩 찾기
for i in input_m:
    start = 0
    end = len(input_n)
    mid = (end + start)//2
    while start <= end:
        mid = (end + start) // 2

        #while문이 끝까지 돌았는데 마지막 값과 i가 같지 않다면 부품이 존재하지 않음
        if start == end and input_n[start] != i:
            print("no", end = ' ')

        #부품이 존재하면 현재 반복문 종료하고 다음 부품 찾기 시작
        if input_n[mid] == i:
            print("yes", end = ' ')
            break
        elif input_n[mid] > i:
            end = mid - 1
        else:
            start = mid + 1
```

```python
#book solution
#이진 탐색 소스코드 구현(반복문)

def binary_search(array, target, start, end):
    while start <= end:
        mid = (start + end) // 2
        
        #찾은 경우 인덱스 반환
        if array[mid] == target:
            return mid
        #중간점의 값보다 찾고자 하는 값이 적은 경우 왼쪽 확인
        elif array[mid] > target:
            end = mid - 1
        #중간점의 값보다 찾고자 하는 값이 작은 경우 오른쪽 확인
        else:
            start = mid + 1
    return None

#N(가게의 부품 개수) 입력
n = int(input())
#가게에 있는 전체 부품 번호를 공백으로 구분하여 입력
array = list(map(int, input().split()))
array.sort() #이진 탐색을 수행하기 위해 사전에 정렬 수행
#M(손님이 확인 요청한 부품 개수) 입력
m = int(input())
#손님이 확인 요청한 전체 부품 번호를 공백으로 구분하여 입력
x = list(map(int, input().split()))

#손님이 확인 요청한 부품 번호를 하나씩 확인
for i in x:
    #해당 부품이 존재하는지 확인
    result = binary_search(array, i, 0, n - 1)
    if result != None:
        print('yes', end = ' ')
    else:
        print('no', end = ' ')
```
