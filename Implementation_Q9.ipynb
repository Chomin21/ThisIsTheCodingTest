```python
#book solution
#step이 핵심

def solution(s):
    answer = len(s)
    #1개 단위(step)부터 압축단위를 늘려가며 확인
    #절반만 확인하는 이유 : anwer의 초깃값이 s. 길어봤자 len(s)임.
    for step in range(1, len(s) // 2 + 1): 
        compressed = ""
        prev = s[0:step] #앞에서부터 step만큼의 문자열 추출
        count = 1
        #단위(step) 크기만큼 증가시키며 이전 문자열과 비교
        for j in range(step, len(s), step): #두번째 요소부터 비교해나감
            #이전 상태와 동일하다면 압축 횟수(count) 증가
            if prev == s[j:j + step]:
                count += 1
            #다른 문자열이 나왔다면(더 이상 압축하지 못하는 경우라면)
            else:
                compressed += str(count) + prev if count >= 2 else prev #1인 경우는 자연스럽게 제외됨
            #만들어지는 압축 문자열이 가장 짧은 것이 정답
            answer = min(answer, len(compressed))
    return answer
    
```
